```python
import requests
import pandas as pd
```

```python
teams_response = requests.get('https://data.nba.net/10s/prod/v2/2022/teams.json').json()
```

```python
teams = teams_response['league']['standard']
```

```python
teams_df = pd.DataFrame(data=teams).set_index('teamId').drop(['isNBAFranchise', 'isAllStar'], axis=1)
```

```python
teams_df
```

```python
box_score_response = requests.get('https://data.nba.net/10s/prod/v1/20220616/0042100406_boxscore.json').json()
```

```python
player_df = pd.DataFrame(box_score_response['stats']['activePlayers']).set_index('personId')
player_df
```

```python
def get_box_score(df, player_name):
	player = df[df['player_code'] == player_name]
	return player.get(['points', 'min', 'fgm', 'fga', 'fgp', 'ftm', 'fta', 'ftp', 'tpm', 'tpa', 'tpp', 'offReb', 'defReb', 'totReb', 'assists', 'pFouls', 'steals', 'turnovers', 'blocks', 'plusMinus'])
```

```python
steph = get_box_score(player_df, 'stephen_curry')
steph
```

```python
klay = get_box_score(player_df, 'klay_thompson')
klay
```
